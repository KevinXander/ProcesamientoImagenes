```

from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

```

```
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/My Drive/Procesamiento Imagenes/ejemplo.jpg')
cv2_imshow(img)
```

# **Usando PLT. ¡las imagenes se pasan en BGR!**

```
plt.figure(figsize=(7,7))
plt.imshow(img)
plt.show()
```

# **Remover Ejes X y Y**

```

plt.figure(figsize=(7,7))
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show
```

# **Cambiar de BGR a RGB**

```
img_RGB=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(img_RGB)
plt.xticks([])
plt.yticks([])
plt.show
```

# **Convertir a escala de grises**

```
img_GRAY=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,10))
plt.imshow(img_GRAY, cmap='gray') #Ojo, debe mapearse la imagen a gris para que salga correctamente en escala de grises
plt.xticks([])
plt.yticks([])
plt.show
```
